```python
#While迴圈(不知道會執行幾次)
#以 random內建函式
#要先import進來
import random
while random.randint(1,10) !=5:
    print("這不是5")
print("這是5")
```

```python
#While Ture:(單向執行)
while True:
    rand_value=random.randint(1,10)
    if rand_value==5:
        print("這是5")
        break
    print("這不是5")
```
