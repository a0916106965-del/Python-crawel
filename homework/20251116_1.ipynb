```python
#字串和變數的整合(範例一)
apple=5
orange=9
print("蘋果="+str(apple)+"個,橘子="+str(orange)+"個")
```

```python
#str template
#字串插補變數
#字串和變數整合,傳出1個全新的字串
#f=format格式化(操作方式為「格式化字串.format(資料)」，輸出結果會將資料插入格式化字串的位置)
#{}=集合,無順序的資料集合
print(f"蘋果={apple}個,橘子={orange}個")
```

```python
#input()呼叫,內建的funtion
#可以沒有參數
#input(prompt(參數,只能1個,使用引數位置的呼叫))
#呼叫稱引數值
#定義稱參數
apple=int(input("請輸入蘋果的數量:"))
print(apple)
type(apple)
```
